```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
df=pd.read_csv("Carbon Emission.csv")
```

```python
# Dropping missing values
df.dropna
```

```python
plt.figure(figsize=(8,3))
sns.histplot(df['CarbonEmission'], bins=10, kde=True)
plt.title('Distribution of Carbon Emissions')
plt.xlabel('Carbon Emission')
plt.ylabel('Frequency')
plt.show()
```

### 1) What is the relationship between diet and carbon emissions?

```python
averageEmisisonsbyDiet=df.groupby('Diet')['CarbonEmission'].mean()
print(averageEmisisonsbyDiet)
```

```python
plt.figure(figsize=(8,3))
sns.boxplot(x='Diet', y='CarbonEmission', data=df)
plt.title('carbon emissions by Diet Type')
plt.xlabel('Diet Type')
plt.ylabel('Carbon Emissions')
plt.show()
```

### 2) How does the frequency of air travel impact carbon emissions?

```python
AverageEmssionsbyAirTravel=df.groupby('Frequency of Traveling by Air')['CarbonEmission'].mean()
print(AverageEmssionsbyAirTravel)
```

```python

df["Frequency of Traveling by Air"] = df["Frequency of Traveling by Air"].astype("category")

palette = {
    'very frequently': 'g',
    'frequently': 'b',
    'rarely': 'r',
    'never': '.35'  # Gray
}

sns.violinplot(data=df, x="Frequency of Traveling by Air", y="CarbonEmission",
               split=True, inner="quart", fill=False,
               palette=palette)

plt.title("Distribution of Carbon Emissions by Frequency of Air Travel")
plt.show()
```

### 3) What role does the type of heating energy source play in carbon emissions?

```python
avg_emissions_by_heating_source = df.groupby('Heating Energy Source')['CarbonEmission'].mean()
print(avg_emissions_by_heating_source)
```

```python
import seaborn as sns
sns.set_theme(style="darkgrid")


# Plot the responses for different events and regions
sns.lineplot(x="Heating Energy Source", y="CarbonEmission",
             sns.title()
             data=df)
```
